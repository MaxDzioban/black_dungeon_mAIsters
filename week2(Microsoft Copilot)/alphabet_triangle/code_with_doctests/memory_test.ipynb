```python
import matplotlib.pyplot as plt
import numpy as np
from memory_profiler import memory_usage
import random

def print_triangle1(num):
    whole_triangle = 0
    last_layer = 0
    letter = ord('A')

    while num > whole_triangle:
        last_layer += 1
        whole_triangle += last_layer

    all_space = whole_triangle * 2 - last_layer
    height = int((all_space ** 0.5) + 0.5)

    for i in range(height):
        for _ in range(height - i):
            if _ == height - 1 - i and i != height - 1:
                for k in range(height - _):
                    print(chr(letter), end=" " if k != height - 1 - _ else "")
                    letter += 1
            elif i < height - 1:
                print(" ", end=" ")
        if i == height - 1:
            for k in range(last_layer - (whole_triangle - num)):
                print(chr(letter), end=" " if k != last_layer - (whole_triangle - num) - 1 else "")
                letter += 1
        print()



def print_triangle2(num):
    whole_triangle = 0
    last_layer = 0
    letter = 65

    while True:

        last_layer += 1
        whole_triangle += last_layer
        if num <= whole_triangle:
            break
        
    all_space = whole_triangle * 2 - last_layer
    height = 1

    while True:
        if height * height == all_space:
            break
        height += 1

    for i in range(height):
        for j in range(height - i):
            if j == height - 1 - i and i != height - 1:
                for k in range(height - j):
                    if k == height - 1 - j:
                        print(chr(letter), end="")
                    else:
                        print(chr(letter), end=" ")
                    letter += 1
            elif i < height - 1:
                print(" ", end=" ")
        if i == height - 1:
            for k in range(last_layer - (whole_triangle - num)):
                if k == last_layer - (whole_triangle - num) - 1:
                    print(chr(letter), end="")
                else:
                    print(chr(letter), end=" ")
                letter += 1
        print()


def memory_test():

    # Memory usage container for each input value
    mem_usages_1 = []
    mem_usages_2 = []
    test_range = range(1, 26, 5)
    # Measure memory usage for each input value
    for num in test_range:
        print(num)
        mem_usage = memory_usage((print_triangle1, (num,), {}))
        mem_usages_1.append(mem_usage[-1])  # Take the maximum memory usage from the list
        mem_usage = memory_usage((print_triangle2, (num,), {}))
        mem_usages_2.append(mem_usage[-1])  # Take the maximum memory usage from the list

    # Plot the results
    plt.plot(test_range, mem_usages_1, label = 'Microsoft Copilot')
    plt.plot(test_range, mem_usages_2, label = 'human')
    plt.title('Memory Usage vs. Input Value')
    plt.xlabel('Input Value')
    plt.ylabel('Memory Usage (MB)')
    plt.ylim(min(mem_usages_1+mem_usages_2) - 0.2, max(mem_usages_1+mem_usages_2) + 0.2)
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == '__main__':    
    memory_test()
```
