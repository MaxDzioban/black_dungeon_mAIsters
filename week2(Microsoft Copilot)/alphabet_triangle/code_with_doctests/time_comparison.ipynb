```python
import time
import matplotlib.pyplot as plt

def print_triangle_1(num):
    whole_triangle = 0
    last_layer = 0
    letter = ord('A')

    while num > whole_triangle:
        last_layer += 1
        whole_triangle += last_layer

    all_space = whole_triangle * 2 - last_layer
    height = int((all_space ** 0.5) + 0.5)

    for i in range(height):
        for _ in range(height - i):
            if _ == height - 1 - i and i != height - 1:
                for k in range(height - _):
                    letter += 1
            elif i < height - 1:
                pass
        if i == height - 1:
            for k in range(last_layer - (whole_triangle - num)):
                letter += 1

def print_triangle_2(num):
    whole_triangle = 0
    last_layer = 0
    letter = 65

    while True:
        last_layer += 1
        whole_triangle += last_layer
        if num <= whole_triangle:
            break
        
    all_space = whole_triangle * 2 - last_layer
    height = 1

    while True:
        if height * height == all_space:
            break
        height += 1

    for i in range(height):
        for j in range(height - i):
            if j == height - 1 - i and i != height - 1:
                for k in range(height - j):
                    if k == height - 1 - j:
                        letter += 1
                    else:
                        letter += 1
            elif i < height - 1:
                pass
        if i == height - 1:
            for k in range(last_layer - (whole_triangle - num)):
                if k == last_layer - (whole_triangle - num) - 1:
                    letter += 1
                else:
                    letter += 1

# Test the functions with different inputs and measure the time
inputs = list(range(1, 1000000, 10000))
times_1 = []
times_2 = []

for i in inputs:
    start = time.time()
    print_triangle_1(i)
    end = time.time()
    times_1.append(end - start)

    start = time.time()
    print_triangle_2(i)
    end = time.time()
    times_2.append(end - start)

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(inputs, times_1, label='bot')
plt.plot(inputs, times_2, label='human')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()

```
